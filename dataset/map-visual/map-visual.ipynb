```python
import pandas as pd
import numpy as np
import folium
from IPython.display import display

```

```python

df = pd.DataFrame(
    np.random.randn(1000, 2) / [50, 50] + [37.76, -122.4],
    columns=["lat", "lon"],
)

# Create a folium map
m = folium.Map(
    location=[df['lat'].mean(), df['lon'].mean()],
    zoom_start=12,
    tiles='CartoDB positron',  # Grayscale tile
    control_scale=True
)

for _, row in df.iterrows():
    folium.CircleMarker(
        location=(row['lat'], row['lon']),
        radius=2, 
        color='red',  
        fill=True,
        fill_color='red',
        fill_opacity=0.1
    ).add_to(m)

display(m)
```
